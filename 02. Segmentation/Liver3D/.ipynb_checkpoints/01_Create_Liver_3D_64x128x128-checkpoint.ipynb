# Augmentation~~~~~~~~~

```python
angle_array = []
scale_array = []

aug_count = 0
if do_augment== True :
    aug_count = 10
    exist_same = True

    while exist_same == True :
        angle_array = np.random.randint(low=-40, high=40, size=(aug_count,)) / 10   # 랜덤 앵글
        scale_array = np.random.randint(low=80, high=100, size=(aug_count,)) / 100  # 랜덤 resize 

        dict_angle = {}
        dict_scale = {}
        for i in range(aug_count) :
            dict_angle[str(angle_array[i])] = 1
            dict_scale[str(scale_array[i])] = 1

        ang_cnt = len(dict_angle)
        sca_cnt = len(dict_scale)
        if (ang_cnt == sca_cnt and ang_cnt == aug_count) :
            exist_same = False
            break

    print(angle_array)
    print(scale_array)
```

```python


all_count = len(image_list)
if do_augment== True :
    all_count = len(image_list) * (1 + aug_count)

all_img = np.zeros((all_count,dst_slice, dst_height, dst_width), dtype=np.uint8)  # 3d 데이터 저장할 배열
all_lab = np.zeros((all_count,dst_slice, dst_height, dst_width), dtype=np.uint8) # 3d 데이터 저장할 배열 
print(all_img.shape)
print(all_lab.shape)
```

```python
all_counter = 0
for i in tqdm_notebook(range(len(image_list))):
    image_path = image_list[i]
    label_path = label_list[i]
    patient_id, date = patient_id_date_extract(image_path)

    itk_img= sitk.ReadImage(image_path)
    itk_lab= sitk.ReadImage(label_path)

    img_arr = sitk.GetArrayFromImage(itk_img)
    lab_arr = sitk.GetArrayFromImage(itk_lab)

    img_win = win_scale(img_arr, win_level, win_width, np.uint8, (0, 255))
    
    vol_img = np.zeros((dst_slice, dst_height, dst_width), dtype=np.uint8)  # 3d 데이터 저장할 배열
    vol_leb = np.zeros((dst_slice, dst_height, dst_width), dtype=np.uint8) # 3d 데이터 저장할 배열 
    counter_j = 0
    for j in range(0, dst_slice):
        idx = img_win.shape[0] - j - 1
        if idx >= img_win.shape[0] or idx < 0:      # 만약에 ct가 64장보다 작을 수도 있기 때문에 그럴 경우를 대비하여
            continue
            
        img_temp = img_win[idx] 
        lab_temp = lab_arr[idx]
        
        sum_1 = np.sum(lab_temp)

        if sum_1 > 0:
        
            img_temp = cv2.flip(img_temp, 0) # 영상 뒤집기 -> 확인해보고 할지말지 결정하기~!
            
            if do_equalizeHist == True :
                img_temp = cv2.equalizeHist(img_temp)
    
            # 3d 데이터가 512x512일 경우, 메모리를 너무 많이 잡아먹을 수도 있기 때문.
            re_image = cv2.resize(img_temp, dsize=(dst_width, dst_height), interpolation=cv2.INTER_AREA) #512 x 512를 128 x 128로 바꾸기 
            re_label = cv2.resize(lab_temp, dsize=(dst_width, dst_height), interpolation=cv2.INTER_AREA) #512 x 512를 128 x 128로 바꾸기  
            
            vol_img[counter_j,:,:] = re_image
            vol_leb[counter_j,:,:] = re_label
            
            all_img[all_counter,counter_j,:,:] = re_image
            all_lab[all_counter,counter_j,:,:] = re_label
            
            counter_j += 1
    all_counter += 1
    
    vol_img = np.expand_dims(vol_img, -1)
    vol_leb = np.expand_dims(vol_leb, -1)
    vol_leb = np.concatenate((1 - vol_leb, vol_leb), -1)
    vol_leb = np.array(vol_leb)
    
    if save_result_image == True:
        # 파일 이름 설정             (↓이게 원본이니까 00)
        image_file_name = 'id_{0:04d}-00_{1}x{2}x{3}_{4}_{5}.npy'.format(i + 1, dst_width, dst_height, dst_slice, patient_id, date)
        label_file_name = 'id_{0:04d}-00_{1}x{2}x{3}_{4}_{5}.npy'.format(i + 1, dst_width, dst_height, dst_slice, patient_id, date)
        
        np.save('{0}/{1}'.format(image_save_path, image_file_name), vol_img)
        np.save('{0}/{1}'.format(label_save_path, label_file_name), vol_leb)

    # 만약 어그멘테이션 한다면 똑같이 re-scale 해줘야겠지?
    if do_augment == True :
        for a in range(aug_count): # 앵글 위에서 10개 만들었지? 

            aang = angle_array[a] # 위에서 랜덤 앵글 만들었던거 불러옴
            asca = scale_array[a] # 위에서 랜덤 resize 스케일
            
            vol_img = np.zeros((dst_slice, dst_height, dst_width), dtype=np.uint8)  # 3d 데이터 저장할 배열
            vol_leb = np.zeros((dst_slice, dst_height, dst_width), dtype=np.uint8) # 3d 데이터 저장할 배열 
            counter_j = 0
            for j in range(0, dst_slice) : 
                idx = img_win.shape[0] - j - 1
                if idx >= img_win.shape[0] or  idx < 0:      #만약 ct장수가 64장보다 작을 경우를 위해
                    continue
                
                img_temp = img_win[idx] # win_level, width에 맞게 조정한 image
                lab_temp = lab_arr[idx] # label

                sum_1 = np.sum(lab_temp)

                if sum_1 > 0:
                
                    img_temp = cv2.flip(img_temp, 0)
                    if do_equalizeHist == True : # 히스토그램 평활화 하는부분
                        img_temp = cv2.equalizeHist(img_temp)
    
                    img_temp = RotateScaleImage(img_temp, aang, asca) # 이미지 사이즈랑 앵글 바꾸는 작업
                    lab_temp = RotateScaleImage(lab_temp, aang, asca) # 라벨도 마찬가지
    
                    re_image = cv2.resize(img_temp, dsize=(dst_width, dst_height), interpolation=cv2.INTER_AREA) #학습에 사용하기 위해 다시 128
                    re_label = cv2.resize(lab_temp, dsize=(dst_width, dst_height), interpolation=cv2.INTER_AREA)
    
                    vol_img[counter_j,:,:] = re_image
                    vol_leb[counter_j,:,:] = re_label
            
                    all_img[all_counter,counter_j,:,:] = re_image
                    all_lab[all_counter,counter_j,:,:] = re_label
                    
                    counter_j +=1
            all_counter += 1
            
            vol_img = np.expand_dims(vol_img, -1)
            vol_leb = np.expand_dims(vol_leb, -1)
            vol_leb = np.concatenate((1 - vol_leb, vol_leb), -1)
            vol_leb = np.array(vol_leb)
            
            if save_result_image == True:
                # 이건 어그멘테이션 한거니까 1부터↓ 두글자로 저장하기 위해 02d
                image_file_name = 'id_{0:04d}-{1:02d}_{2}x{3}x{4}_{5}_{6}.npy'.format(i + 1, a + 1, dst_width, dst_height, dst_slice, patient_id, date)
                label_file_name = 'id_{0:04d}-{1:02d}_{2}x{3}x{4}_{5}_{6}.npy'.format(i + 1, a + 1, dst_width, dst_height, dst_slice, patient_id, date)
                
                np.save('{0}/{1}'.format(image_save_path, image_file_name), vol_img)
                np.save('{0}/{1}'.format(label_save_path, label_file_name), vol_leb)
all_img = np.expand_dims(all_img, -1)
all_lab = np.expand_dims(all_lab, -1)
all_lab = np.concatenate((1 - all_lab, all_lab), -1)

all_img = np.array(all_img)            
all_lab = np.array(all_lab)

print(all_img.shape)
print(all_lab.shape)
print(vol_img.shape)
print(vol_leb.shape)
    
endTime = datetime.now()
runTime = endTime - startTime
print('End Time : {0}'.format(endTime))
print('Running Time : ' + str(runTime))

print('Finish save Image & Mask nifti')
```
